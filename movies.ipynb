```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

A lemmatizer is a tool in Natural Language Processing (NLP) that reduces words to their base or dictionary form, called a lemma.
For example:

"running" → "run"

"better" → "good"

"mice" → "mouse"

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text = str(text).lower()
  text = re.sub(r'[^a-zA-Z]', ' ', text)
  words = text.split()
  words = [word for word in words if word not in stop_words]
  words = [lemmatizer.lemmatize(word) for word in words]
  return ' '.join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

## TF-IDF

TF-IDF stands for Term Frequency – Inverse Document Frequency.
It’s a statistical measure used in Natural Language Processing (NLP) to evaluate how important a word is to a document in a collection of documents (called a corpus).

The main idea:

- Words that appear frequently in a document should have higher weight.
- But common words (like “the”, “is”, “and”) that appear in many documents should be penalized (given less weight).
## Calculation

It has two parts:

1. Term Frequency (TF)
Measures how often a word appears in a single document.

$\text{TF}(t, d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of words in document } d}$

Example: If “AI” appears 3 times in a 100-word document → TF = 3/100 = 0.03

2. Inverse Document Frequency (IDF)
Measures how rare a word is across all documents in the corpus.

$\text{IDF}(t, D) = \log \left( \frac{\text{Total number of documents in corpus } D}{\text{Number of documents containing term } t} \right)$

Example: If “AI” appears in 10 out of 1000 documents →
IDF = log(1000 / 10) = log(100) ≈ 2

### TF-IDF Score

$\text{TF-IDF}(t, d, D) = \text{TF}(t, d) \times \text{IDF}(t, D)$

High TF-IDF = word is common in this document but rare in others → likely important!

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

### N-Grams

n-grams are contiguous sequences of n items (usually words or characters) from a given text or speech sample. They are widely used in Natural Language Processing (NLP) for tasks like language modeling, text generation, machine translation, and more.

- Unigrams (n=1): Single words.

  Example: "I love NLP" → ["I", "love", "NLP"]

- Bigrams (n=2): Pairs of adjacent words.
  
  Example: "I love NLP" → ["I love", "love NLP"]

- Trigrams (n=3): Triplets of adjacent words.

  Example: "I love NLP" → ["I love NLP"]

```python
tfidf = TfidfVectorizer(max_features=50000, stop_words = 'english', ngram_range = (1, 2))
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

## Cosine Similarity

Cosine similarity is a measure of how similar two non-zero vectors are, regardless of their magnitude. In NLP, it’s commonly used to compare text documents or word embeddings by measuring the cosine of the angle between their vector representations.

The result ranges from -1 to 1:

- 1 → vectors point in the exact same direction (identical in orientation).
- 0 → vectors are orthogonal (no similarity).
- -1 → vectors point in opposite directions.

$\text{Cosine Similarity} = \frac{A \cdot B}{\|A\| \|B\|} = \frac{\sum_{i=1}^{n} A_i B_i}{\sqrt{\sum_{i=1}^{n} A_i^2} \sqrt{\sum_{i=1}^{n} B_i^2}}$

For detail understanding of Cosine Similarity, Checkout [this](http://medium.com/data-science-collective/cosine-similarity-explained-the-math-behind-llms-b20caac9f93c).

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n=10):
    # Check if the requested movie title exists in our index mapping
    if title not in indices:
        return ['Movie not found']

    # Get the row index (integer position) of the movie in the DataFrame
    idx = indices[title]

    # Compute cosine similarity between the target movie's TF-IDF vector
    # and all other movies' TF-IDF vectors.
    # Result: a 2D array with shape (1, num_movies)
    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    # .flatten() converts it to a 1D array for easier handling

    # Get indices of movies sorted by similarity (highest to lowest):
    # - argsort() returns indices that would sort the array in ascending order
    # - [::-1] reverses to get descending order (most similar first)
    # - [1:n+1] skips the first result (which is the movie itself) and takes next `n` movies
    similar_idx = sim_score.argsort()[::-1][1:n+1]

    # Use iloc to fetch the titles of the top-n most similar movies
    return df['title'].iloc[similar_idx].tolist()  # Optional: .tolist() for clean list output
```

```python
recommend('Avatar')
```
